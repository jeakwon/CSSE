```python
from csse.utils.experiment import Load_ResNet18_CIFAR100_CIL_Experiment
```

```python
import pandas as pd
dfs = []
for algo in ['bp', 'cb', 'rt']:
    for seed in [0, 1, 2]:
        exp = Load_ResNet18_CIFAR100_CIL_Experiment(algo=algo, seed=seed, sessions=range(21), device='cuda', batch_size=100, num_workers=2)
        df = pd.DataFrame(exp.test_accs())
        df.to_csv(f'test_accs_{algo}_{seed}.csv', index=False)

        dfs.append(df)

df = pd.concat(dfs)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


fig, ax = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
plt.subplots_adjust(wspace=0.1)
sns.lineplot(data=df, x='session', y='all_classes', hue='algo', ax=ax[0])
ax[0].set_ylabel('Test Acc')
ax[0].set_title('All Classes')
sns.lineplot(data=df, x='session', y='old_classes', hue='algo', ax=ax[1])
ax[1].set_ylabel('Test Acc')
ax[1].set_title('Old Classes')
sns.lineplot(data=df, x='session', y='new_classes', hue='algo', ax=ax[2])
ax[2].set_ylabel('Test Acc')
ax[2].set_title('New Classes')
```

```python
tmp = df.copy()

for algo in ['bp', 'cb', 'rt']:
    mask = tmp['algo'] == algo
    ref_mask = tmp['algo'] == 'rt'

    tmp.loc[mask, ['all_classes', 'old_classes', 'new_classes']] = (
        tmp.loc[mask, ['all_classes', 'old_classes', 'new_classes']].values -
        tmp.loc[ref_mask, ['all_classes', 'old_classes', 'new_classes']].values
    )

fig, ax = plt.subplots(1, 3, figsize=(15, 4), sharey=True)
plt.subplots_adjust(wspace=0.1)
sns.lineplot(data=tmp, x='session', y='all_classes', hue='algo', ax=ax[0])
ax[0].set_ylabel('Test Acc')
ax[0].set_title('All Classes')
sns.lineplot(data=tmp, x='session', y='old_classes', hue='algo', ax=ax[1])
ax[1].set_ylabel('Test Acc')
ax[1].set_title('Old Classes')
sns.lineplot(data=tmp, x='session', y='new_classes', hue='algo', ax=ax[2])
ax[2].set_ylabel('Test Acc')
ax[2].set_title('New Classes')
```
